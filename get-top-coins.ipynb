```python
import requests
import datetime
import pandas as pd
import numpy as np
```

```python
#define functions

def info(number_of_coins=100):
    url = 'https://api.coinmarketcap.com/v1/ticker/?limit={}'.format(number_of_coins)
    page = requests.get(url)
    data = page.json()
    df = pd.DataFrame.from_dict(data)
    df['last_update_date'] = [datetime.datetime.fromtimestamp(int(d)) for d in df.last_updated]
    return df
```

```python
info = info(200)
info.columns.values
```

```python
info.head()
```

```python
top_coins = info[['symbol','name', 'market_cap_usd','last_update_date']]
```

```python
#save info about market capitalization
top_coins.to_csv('top-coins-info.csv', index = False)
```

```python
coins_list = list(top_coins.symbol)

#some coins have different symbols at cryptocompare.com and coinmarketcap.com
additional_list = ['IOT', 'BTM*']
coins_list.extend(additional_list)
```
